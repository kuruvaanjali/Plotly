## What is Plotly?
Plotly is a powerful, interactive graphing library for Python. It’s widely used for creating dynamic, web-based visualizations such as line charts, bar graphs, scatter plots, heatmaps, and even advanced dashboards.

It supports both 2D and 3D plots.

One of its biggest strengths is interactivity — you can zoom, pan, hover, and export visuals easily.

## What is Cufflinks?
Cufflinks bridges the gap between Pandas and Plotly. It enables you to plot directly from Pandas dataframes using a simple .iplot() function.

Makes interactive plotting quick and easy.

Best used when working with Pandas dataframes.

Works great in Jupyter environments.

## Parallel Coordinates Plot

```python
import plotly.express as px

fig = px.parallel_coordinates(
    df,  # Data source
    dimensions=["Unit price", "Quantity", "Tax 5%"],  # Selected numerical features for axes
    color="Unit price",  # Color mapping based on Unit Price
    color_continuous_scale=px.colors.diverging.Tealrose,  # Color scale for better contrast
    labels={"Unit price": "Unit Price", "Quantity": "Quantity", "Tax 5%": "Tax (%)"}  # Custom axis labels
)

fig.update_layout(
    title={
        "text": "Parallel Coordinates Plot of Sales Data",
        "y": 0,  # Position title at the bottom
        "x": 0.5,
        "xanchor": "center",
        "yanchor": "bottom"
    }
)

fig.show()  # Display the plot
```

### Summary
This plot visualizes relationships between Unit Price, Quantity, and Tax 5% across all transactions. Each line represents a single transaction, helping us spot patterns and correlations between variables. The color gradient (Tealrose) reflects the Unit Price, aiding quick comparison.

## Sunburst Chart

```python
import plotly.express as px

# Create a new column 'Total' by multiplying Quantity and Unit price
df["Total"] = df["Quantity"] * df["Unit price"]

# Generating a sunburst chart to visualize Total sales distribution across City and Product line
fig = px.sunburst(df, 
                  path=["City", "Product line"],  # Hierarchical levels
                  values="Total",                 # Size of segments
                  color="City")                   # Color coding by city

# Add title at the bottom of the chart
fig.update_layout(
    title={
        "text": "Sunburst Chart of Total Sales by City and Product Line",
        "y": 0,  # Bottom position
        "x": 0.5,
        "xanchor": "center",
        "yanchor": "bottom"
    }
)

fig.show()
```

### Sunburst Chart Showing Sales Distribution by City and Product Line

### Summary:
The above sunburst chart visually represents the total sales (Total) broken down by City and further by each Product line. This hierarchical view helps in quickly understanding which product categories contribute the most within each city. Each segment's size correlates with the sales volume, and the colors differentiate between the cities.

## Treemap

```python
import plotly.express as px

fig = px.treemap(
    df, 
    path=["City", "Product line"], 
    values="Total", 
    color="Total", 
    hover_data=["Gender"]
)

# Add a title at the top of the chart
fig.update_layout(
    title={
        "text": "Treemap of Total Sales by City and Product Line",
        "y": 0.95,
        "x": 0.5,
        "xanchor": "center",
        "yanchor": "top"
    }
)

fig.show()
```

#### Treemap Visualization of Sales by City and Product Line

### Summary:
This treemap chart offers a clear and compact visual breakdown of total sales (Total) categorized by City and Product line. The size of each rectangle reflects the sales volume, while the color gradient (from dark to bright) further emphasizes the magnitude of sales. Hovering over each block reveals the Gender data for more insights.

## Heatmap

```python
import plotly.graph_objects as go
import numpy as np

corr = df.corr(numeric_only=True)  # Gets correlation only between numerical columns

fig = go.Figure(data=go.Heatmap(
    z=corr.values,               # The correlation matrix values
    x=corr.columns,              # Labels for x-axis
    y=corr.columns,              # Labels for y-axis
    colorscale="Blues"           # Blue color gradient
))

# Add a title at the top of the heatmap
fig.update_layout(
    title={
        "text": "Correlation Heatmap of Numerical Features",
        "y": 0.95,
        "x": 0.5,
        "xanchor": "center",
        "yanchor": "top"
    }
)

fig.show()
```

### Summary:
This heatmap visually represents the correlation between numerical variables like Unit price, Quantity, Tax 5%, and Total in the dataset.
The color intensity indicates the strength of correlation — darker blue means stronger correlation (either positive or negative).
It helps identify relationships between variables, which can guide further analysis or feature selection.

## Animation Plot

```python
# creating animated bar chart
fig = px.bar(df, 
             x="Product line", 
             y="Total", 
             color="Product line",
             animation_frame="Branch",
             title="Sales by Product Line across Branches",
             labels={"Total": "Total Sales"},
             height=500)

fig.update_layout(xaxis={'categoryorder':'total descending'})
fig.show()
```

### Summary:
This animated bar chart visualizes how total sales vary across different product lines for each branch (A, B, and C).
The animation provides an intuitive way to compare performance branch-wise in real-time, making it easier to spot patterns and high-performing product categories.
